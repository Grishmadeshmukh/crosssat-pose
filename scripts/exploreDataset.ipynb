## Inspect one satellite

```python
from pathlib import Path
import json
import zipfile
import os

# --------------------------------------------------
# CONFIG
# --------------------------------------------------
SATELLITE_NAME = "acrimsat_final"   # change this if needed

# Robust path handling for notebook
DATA_DIR = Path("data")
if not DATA_DIR.exists():
    DATA_DIR = Path("../data")

SAT_DIR = DATA_DIR / SATELLITE_NAME

print("Current working directory:", os.getcwd())
print("Using data dir:", DATA_DIR.resolve())
print("Inspecting satellite:", SAT_DIR.resolve())
print("Exists:", SAT_DIR.exists())

if not SAT_DIR.exists():
    raise FileNotFoundError(f"Satellite folder not found: {SAT_DIR}")

# --------------------------------------------------
# 1) LABELS.JSON
# --------------------------------------------------
labels_path = SAT_DIR / "labels.json"

print("\n" + "="*60)
print("LABELS.JSON INFO")
print("="*60)

if labels_path.exists():
    with open(labels_path, "r") as f:
        labels = json.load(f)

    print("labels.json exists: YES")
    print("Type:", type(labels))

    if isinstance(labels, list):
        print("Number of entries:", len(labels))
        if len(labels) > 0:
            print("\nFirst entry keys:")
            print(list(labels[0].keys()))

            print("\nFirst entry values:")
            for k, v in labels[0].items():
                print(f"{k}: {v}")

    elif isinstance(labels, dict):
        print("Top-level keys:")
        print(list(labels.keys())[:20])

        # show one example entry if possible
        keys = list(labels.keys())
        if len(keys) > 0:
            example_key = keys[0]
            print("\nExample key:")
            print(example_key)

            print("\nExample value:")
            print(labels[example_key])

else:
    print("labels.json exists: NO")

# --------------------------------------------------
# 2) IMAGES.ZIP
# --------------------------------------------------
images_zip_path = SAT_DIR / "images.zip"

print("\n" + "="*60)
print("IMAGES.ZIP INFO")
print("="*60)

if images_zip_path.exists():
    with zipfile.ZipFile(images_zip_path, "r") as z:
        image_files = [n for n in z.namelist() if not n.endswith("/")]

        print("images.zip exists: YES")
        print("Number of files inside:", len(image_files))

        print("\nFirst 20 image files:")
        for name in image_files[:20]:
            print(name)

        # inspect one image file size
        if len(image_files) > 0:
            info = z.getinfo(image_files[0])
            print("\nExample file info:")
            print("Filename:", info.filename)
            print("Compressed size:", info.compress_size)
            print("Original size:", info.file_size)
else:
    print("images.zip exists: NO")

# --------------------------------------------------
# 3) MASKS.ZIP
# --------------------------------------------------
masks_zip_path = SAT_DIR / "masks.zip"

print("\n" + "="*60)
print("MASKS.ZIP INFO")
print("="*60)

if masks_zip_path.exists():
    with zipfile.ZipFile(masks_zip_path, "r") as z:
        mask_files = [n for n in z.namelist() if not n.endswith("/")]

        print("masks.zip exists: YES")
        print("Number of files inside:", len(mask_files))

        print("\nFirst 20 mask files:")
        for name in mask_files[:20]:
            print(name)

        if len(mask_files) > 0:
            info = z.getinfo(mask_files[0])
            print("\nExample file info:")
            print("Filename:", info.filename)
            print("Compressed size:", info.compress_size)
            print("Original size:", info.file_size)
else:
    print("masks.zip exists: NO")

# --------------------------------------------------
# 4) MODELS/
# --------------------------------------------------
models_dir = SAT_DIR / "models"

print("\n" + "="*60)
print("MODELS DIRECTORY INFO")
print("="*60)

if models_dir.exists():
    model_files = [p for p in models_dir.rglob("*") if p.is_file()]

    print("models/ exists: YES")
    print("Number of files:", len(model_files))

    print("\nFirst 20 model files:")
    for f in model_files[:20]:
        print(f.relative_to(SAT_DIR))
else:
    print("models/ exists: NO")

print("\n" + "="*60)
print("DONE")
print("="*60)
```
